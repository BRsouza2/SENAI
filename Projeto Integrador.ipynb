```python
#Demonstração inicial das forças

# Definindo as dimensões do carrinho
carrinho_comprimento = 1400  # mm
carrinho_altura = 30  # mm
ponto_A = 100  # mm da extremidade esquerda
ponto_B = 1268  # mm da extremidade esquerda
centro_banco = 1010  # mm da extremidade esquerda
centro_estrutura = 700  # mm da extremidade esquerda

# Criando a figura e os eixos com nova altura
fig, ax = plt.subplots(figsize=(14, 2))

# Desenhando o carrinho como um retângulo
carrinho = patches.Rectangle((0, 0), carrinho_comprimento, carrinho_altura, linewidth=1, edgecolor='black', facecolor='lightgrey')
ax.add_patch(carrinho)

# Marcando os pontos A e B com as anotações
plt.plot([ponto_A, ponto_A], [-10, carrinho_altura + 10], color='red', linestyle='--')
plt.text(ponto_A, -25, 'Ponto A (100 mm)', color='red', fontsize=10, ha='center')

plt.plot([ponto_B, ponto_B], [-10, carrinho_altura + 10], color='blue', linestyle='--')
plt.text(ponto_B, -25, 'Ponto B (1268 mm)', color='blue', fontsize=10, ha='center')

# Marcando o centro de massa do banco com a anotação
plt.plot([centro_banco, centro_banco], [-10, carrinho_altura + 10], color='green', linestyle='--')
plt.text(centro_banco, -25, 'Meio do banco (1010 mm)', color='green', fontsize=10, ha='center')

# Marcando o centro de massa da estrutura com a anotação
plt.plot([centro_estrutura, centro_estrutura], [-10, carrinho_altura + 10], color='purple', linestyle='--')
plt.text(centro_estrutura, -25, 'Meio da estrutura (700 mm)', color='purple', fontsize=10, ha='center')

# Adicionando as setas das forças
plt.arrow(ponto_A, 20, 0, 10, head_width=20, head_length=5, fc='red', ec='red')
plt.arrow(ponto_B, 20, 0, 10, head_width=20, head_length=5, fc='blue', ec='blue')
plt.arrow(centro_banco, carrinho_altura + 20, 0, -10, head_width=20, head_length=5, fc='green', ec='green')
plt.arrow(centro_estrutura, carrinho_altura + 20, 0, -10, head_width=20, head_length=5, fc='purple', ec='purple')

# Adicionando as anotações das forças com os valores
plt.text(ponto_A - 40, carrinho_altura + 35, '$F_{AY}$', color='red', fontsize=12)
plt.text(ponto_B - 40, carrinho_altura + 35, '$F_{BY}$', color='blue', fontsize=12)
plt.text(centro_banco - 60, carrinho_altura + 35, '$P_{banco}$ = 550N', color='green', fontsize=12, ha='center')
plt.text(centro_estrutura - 60, carrinho_altura + 35, '$P_{est}$ = 105N', color='purple', fontsize=12, ha='center')

# Ajustando os limites dos eixos
plt.xlim(-100, carrinho_comprimento + 100)
plt.ylim(-40, carrinho_altura + 50)

# Ajustando o aspecto do gráfico
plt.gca().set_aspect('equal', adjustable='box')
plt.axis('off')

# Mostrando o gráfico
plt.show()
```
